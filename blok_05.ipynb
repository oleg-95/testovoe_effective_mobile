### Задание 3: Ручной счёт корреляции
Рассчитайте линейную корреляцию Пирсона, на основе данных.  
Какой вывод можно сделать на основе полученного результата? Можно ли утверждать, что существует причинно-следственная связь между количеством чашек кофе, выпитых студентами в течение экзаменационного дня, и их итоговым баллом за экзамен?
| Число выпитых чашек кофе | Балл за экзамен |
|-------------------------|-----------------|
| 1 | 85 |
| 1 | 88 |
| 2 | 79 |
| 2 | 81 |
| 2 | 84 |
| 2 | 65 |
| 3 | 67 |
| 3 | 58 |
| 3 | 76 |
| 4 | 49 |

```python
from scipy import stats
import pandas as pd

cofe = pd.Series([1, 1, 2, 2, 2, 2, 3, 3, 3, 4])
score = pd.Series([85, 88, 79, 81, 84, 65, 67, 58, 76, 49])

corr = stats.pearsonr(cofe, score).statistic
print(corr.round(2))
```

#### Ответ:
Корреляция сильная отрицательная, чем больше чашек кофе - тем ниже балл. НО! Корреляция НЕ отражает причинно-следвственную связь между любыми переменными, в данном случае нельзя сказать, что выпитые чашки кофе снижают баллы за экзамен.
